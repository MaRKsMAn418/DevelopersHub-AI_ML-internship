# Task 1: Exploring and Visualizing the Iris Dataset

**Objective:** Load, inspect, and visualize the Iris dataset to understand data trends and distributions.

**Dataset:** Iris Dataset (loaded via seaborn)

## 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a consistent visual style
sns.set_theme(style='whitegrid', palette='Set2')
plt.rcParams['figure.figsize'] = (8, 5)

print('Libraries loaded successfully.')
```

## 2. Load the Dataset

```python
# Load Iris dataset directly from seaborn
df = sns.load_dataset('iris')

print('Dataset loaded.')
print(f'Shape: {df.shape}')          # rows x columns
print(f'Columns: {list(df.columns)}')
```

## 3. Initial Inspection

```python
# First few rows
print(df.head())
```

```python
# Data types and non-null counts
print(df.info())
```

```python
# Descriptive statistics for numerical features
print(df.describe().round(2))
```

```python
# Check for missing values
print('Missing values per column:')
print(df.isnull().sum())
```

## 4. Visualizations

### 4.1 Scatter Plot – Feature Relationships

```python
# Pairplot to show relationships between all feature pairs, coloured by species
sns.pairplot(df, hue='species')
plt.suptitle('Feature Relationships by Species', y=1.02)
plt.tight_layout()
plt.show()
```

### 4.2 Histograms – Value Distributions

```python
df.hist(bins=20, figsize=(10, 8))
plt.suptitle('Feature Value Distributions by Species')
plt.tight_layout()
plt.show()
```

### 4.3 Box Plots – Outlier Detection

```python
sns.boxplot(data=df)
plt.title("Box Plots: Feature Spread and Outliers")
plt.tight_layout()
plt.show()
```
