## 2. Import Libraries and Configure API

```python
from groq import Groq
import re

# -------------------------------------------------------
# PASTE YOUR GROQ API KEY HERE
API_KEY = "gsk_XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX"
# -------------------------------------------------------

client = Groq(api_key=API_KEY)

print('Configuration done.')
```

## 3. System Prompt Design (Prompt Engineering)

```python
SYSTEM_PROMPT = """You are a friendly and knowledgeable general health assistant.
Your role is to provide clear, accurate, and easy-to-understand answers to general health questions.

Important guidelines you must always follow:
1. Answer only general health and wellness questions (symptoms, medications, nutrition, hygiene, first aid).
2. NEVER diagnose any medical condition. Always remind the user to consult a qualified doctor for diagnosis or treatment decisions.
3. NEVER recommend specific prescription medications or dosages.
4. Keep your response concise – ideally 3 to 5 sentences.
5. Use simple, non-technical language so anyone can understand.
6. If a question is outside general health (e.g., surgery details, mental health crisis), gently redirect the user to seek professional help.
7. Always maintain an empathetic and supportive tone."""

print('System prompt set.')
```

## 4. Safety Filter

```python
BLOCKED_PATTERNS = [
    r'\bsuicid\w*\b',
    r'\bself.harm\b',
    r'\boverdos\w*\b',
    r'\bprescri\w*\b',
    r'\bkill myself\b',
    r'\billegal drug\w*\b',
    r'\bdrug abuse\b',
]

SAFETY_RESPONSE = (
    "I'm sorry, but that topic falls outside what I'm able to help with. "
    "Please reach out to a qualified healthcare professional or, in an emergency, "
    "contact your local emergency services immediately."
)

def is_safe_query(query: str) -> bool:
    """Return False if the query matches any blocked pattern."""
    query_lower = query.lower()
    for pattern in BLOCKED_PATTERNS:
        if re.search(pattern, query_lower):
            return False
    return True

print('Safety filter loaded.')
```

## 5. Chatbot Function

```python
def query_health_chatbot(user_query: str) -> str:
    """
    Main function: applies safety filter, calls Groq API,
    and returns the assistant's response.
    """
    # Step 1 – Safety check
    if not is_safe_query(user_query):
        return SAFETY_RESPONSE

    # Step 2 – Call Groq API
    try:
        response = client.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[
                {"role": "system", "content": SYSTEM_PROMPT},
                {"role": "user",   "content": user_query}
            ],
            max_tokens=300,
            temperature=0.7
        )
        return response.choices[0].message.content.strip()

    except Exception as e:
        return f"An unexpected error occurred: {e}"


print('Chatbot function ready.')
```

## 6. Test with Example Queries

```python
test_queries = [
    "What causes a sore throat?",
    "Is paracetamol safe for children?",
    "How much water should I drink per day?",
    "What are common symptoms of dehydration?",
    "How can I treat a mild headache at home?",
    # Edge case – should trigger safety filter
    "What is the overdose limit for paracetamol?",
]

for query in test_queries:
    print(f"{'='*65}")
    print(f"USER: {query}")
    print(f"{'─'*65}")
    response = query_health_chatbot(query)
    print(f"ASSISTANT: {response}")
    print()
```

## 7. Interactive Chat Loop

Run this cell to chat with the bot interactively. Type `quit` to exit.

```python
print("Health Assistant Chatbot")
print("Type your health question below. Type 'quit' to exit.")
print("=" * 50)

while True:
    user_input = input("\nYou: ").strip()

    if not user_input:
        continue
    if user_input.lower() in ('quit', 'exit', 'q'):
        print("Goodbye! Stay healthy.")
        break

    reply = query_health_chatbot(user_input)
    print(f"\nAssistant: {reply}")
```
